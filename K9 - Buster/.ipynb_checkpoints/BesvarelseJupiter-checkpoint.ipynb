Her bruges formlen: $\displaystyle n \geq \bigg(\frac{z_{\alpha/2}\cdot \sigma}{B}\bigg)^2$

Hvor $B$ er den nskede bredde for konfidensintervallet. <br>

```r
alfa = 0.05
B = 0.05
sigma = 0.1
z_alfa_halve = qnorm(1-alfa/2)
n = ((z_alfa_halve * sigma)/(B))**2
n = ceiling(n)
n
```

Rundet op til nrmeste heltal, giver det alts: <br>
$n = 16$

Vi har en forventning om at $\sigma^2$ er strre end 0.1, derfor er det vores alternativ hypotese. Vores nulhypoteste er det modsatte. Her er $\sigma_0^2 = 0.1$

$H_0: \sigma^2 = \sigma_0^2$ <br>
$H_1: \sigma^2 > \sigma_0^2$

Dette er en chi i anden fordeling med n-1 frihedsgrader.

```r
sigma = 0.1
D = read.table("Data_fjederkonstant.csv", header=TRUE, sep=";")
fjeder_konstant = D$Fjederkonstant
s = sd(fjeder_konstant)
n = length(fjeder_konstant)

# Beregn teststrrelse
chi0 = ((n-1)*s**2)/(sigma**2)
chi0

# Beregn kritisk region
# Hjrehalet
alfa = 0.05
chi_alfa = qchisq(1-alfa, n-1)
chi_alfa
```

Da $\sigma_0^2 > \sigma_{\alpha}^2$, forkastes nulhypotesen. Det kan alts bevises p 5% signifikansniveau at variansen er strre end 0.1.

```r
chi_hojre = qchisq(1-alfa/2, n-1)
chi_venstre = qchisq(alfa/2, n-1)
chi_hojre
chi_venstre
```

```r
KI_lav = ((n-1)*s**2)/chi_hojre
KI_hoj = ((n-1)*s**2)/chi_venstre
KI_lav
KI_hoj
```

95% konfidensintervallet for $\sigma^2$ er alts $[0.017; 0.061]$. I dette er 0.1 slet ikke inkluderet, men lavere end intervallet, og derfor er det meget sandsynligt at variansen er hjere end 0.1.

Det er en hypotesetest om middelvrdien hvor populationsvariansen kendes. Derfor bruges flgende:


Dette er en normalfordeling, og kaldet z-fordeling.

Venstrehalet. Derfor forkastes nulhypotesen hvis $z_0 < z_{\alpha}$

```r
# Beregn teststrrelse
mu0 = 4.0
x_streg = mean(fjeder_konstant)
z0 = (x_streg-mu0)/(sigma/sqrt(n))
z0
```
