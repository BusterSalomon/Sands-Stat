Her bruges formlen: $\displaystyle n \geq \bigg(\frac{z_{\alpha/2}\cdot \sigma}{B}\bigg)^2$

Hvor $B$ er den ønskede bredde for konfidensintervallet. <br>

```r
alfa = 0.05
B = 0.05
sigma = 0.1
z_alfa_halve = qnorm(1-alfa/2)
n = ((z_alfa_halve * sigma)/(B))**2
n = ceiling(n)
n
```

Rundet op til nærmeste heltal, giver det altså: <br>
$n = 16$

Vi har en forventning om at $\sigma^2$ er større end 0.1, derfor er det vores alternativ hypotese. Vores nulhypoteste er det modsatte. Her er $\sigma_0^2 = 0.1$

$H_0: \sigma^2 = \sigma_0^2$ <br>
$H_1: \sigma^2 > \sigma_0^2$

Dette er en chi i anden fordeling med n-1 frihedsgrader.

```r
sigma = 0.1
D = read.table("Data_fjederkonstant.csv", header=TRUE, sep=";")
fjeder_konstant = D$Fjederkonstant
s = sd(fjeder_konstant)
n = length(fjeder_konstant)

# Beregn teststørrelse
chi0 = ((n-1)*s**2)/(sigma**2)
chi0

# Beregn kritisk region
# Højrehalet
alfa = 0.05
chi_alfa = qchisq(1-alfa, n-1)
chi_alfa
```

Da $\sigma_0^2 > \sigma_{\alpha}^2$, forkastes nulhypotesen. Det kan altså bevises på 5% signifikansniveau at variansen er større end 0.1.

```r
chi_hojre = qchisq(1-alfa/2, n-1)
chi_venstre = qchisq(alfa/2, n-1)
chi_hojre
chi_venstre
```

```r
KI_lav = ((n-1)*s**2)/chi_hojre
KI_hoj = ((n-1)*s**2)/chi_venstre
KI_lav
KI_hoj
```

95% konfidensintervallet for $\sigma^2$ er altså $[0.017; 0.061]$. I dette er 0.1 slet ikke inkluderet, men lavere end intervallet, og derfor er det meget sandsynligt at variansen er højere end 0.1.

Det er en hypotesetest om middelværdien hvor populationsvariansen kendes. Derfor bruges følgende:


Dette er en normalfordeling, og kaldet z-fordeling.

Venstrehalet. Derfor forkastes nulhypotesen hvis $z_0 < z_{\alpha}$

```r
# Beregn teststørrelse
mu0 = 4.0
x_streg = mean(fjeder_konstant)
z0 = (x_streg-mu0)/(sigma/sqrt(n))
z0
```

```r
# Beregn kritisk region
alfa = 0.05
z_alfa = qnorm(alfa)
z_alfa
```

Da $z_0 < z_{\alpha}$ forkastes nulhypotesen. Dvs. at det kan konkluderes indenfor 5% signifikansniveau at middelværdien for fjedrenes fjederkonstant er mindre end 4.0

Her har vi antaget CSG (centrale grænseværdi sætning). Denne er mildere end hypotesetesten for varians, hvor det blev antaget at stikprøven kom fra en normalfordeling.
